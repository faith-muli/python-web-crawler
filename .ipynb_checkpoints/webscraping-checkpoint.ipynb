```python
import requests;
from bs4 import BeautifulSoup;


r=requests.get("http://www.century21.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/");

c=r.content


soup=BeautifulSoup(c, "html.parser")



all= soup.find_all("div", {"class":"property-card-primary-info"});

len(all)


all[0].find("a",{"class":"listing-price"}).text.replace("\n", "").replace(" ", "")

```

```python
l=[]
for item in all:
    d={}
    d["Price"]=item.find("a", {"class":"listing-price"}).text.replace("\n", "").replace(" ","");
    d["Address"]=item.find_all("div", {"class":"property-address-info"})[0].text.replace("\n", "").replace(" ","");
   
    try:
       d["Beds"]=item.find("div", {"class":"property-beds"}).find("strong").text
    except:
        d["Beds"]="none"
    try:
        d["Baths"]=item.find("div", {"class":"property-baths"}).find("strong").text
    except:
        d["Baths"]="no baths"
    try:
        d["Size"]=item.find("div", {"class":"property-sqft"}).find("strong").text
    except:
        d["Size"]="not available"
    l.append(d)

   
    
```

```python
l
```

```python
import pandas
df=pandas.DataFrame(l)
df
df.to_csv("property.csv")
```
